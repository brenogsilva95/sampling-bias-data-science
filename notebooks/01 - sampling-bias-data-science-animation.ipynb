```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.patches import Rectangle
from IPython.display import HTML

np.random.seed(42)
n = 600

x_a = np.random.normal(4.2, 1.2, n // 2)
y_a = np.random.normal(5.0, 1.4, n // 2)

x_b = np.random.normal(6.5, 1.3, n // 2)
y_b = np.random.normal(4.2, 1.2, n // 2)

X = np.concatenate([x_a, x_b])
Y = np.concatenate([y_a, y_b])
grp = np.array(["Grupo A"] * (n // 2) + ["Grupo B"] * (n // 2))

sample_mask = (X > 5.2) & (Y > 3.5) & (Y < 6.8)

pop_a = np.mean(grp == "Grupo A")
pop_b = np.mean(grp == "Grupo B")

sample_a = np.mean(grp[sample_mask] == "Grupo A")
sample_b = np.mean(grp[sample_mask] == "Grupo B")

xmin, xmax = 0, 11
ymin, ymax = 0, 10

fig, ax = plt.subplots(figsize=(8, 8))
fig.patch.set_facecolor("white")

title_main = "O problema nem sempre está no modelo"
subtitle = "Às vezes, está no dado que chegou até ele"

def update(frame):
    ax.clear()
    ax.set_facecolor("white")
    ax.set_xlim(xmin, xmax)
    ax.set_ylim(ymin, ymax)
    ax.set_xticks([])
    ax.set_yticks([])

    for spine in ax.spines.values():
        spine.set_visible(False)

    ax.set_title(title_main, fontsize=18, pad=28, weight="bold")
    ax.text(
        0.5, 1.00, subtitle,
        transform=ax.transAxes,
        ha="center", va="bottom",
        fontsize=11
    )

    if frame < 35:
        alpha = min(1, 0.25 + frame / 40)

        ax.scatter(
            X[grp == "Grupo A"], Y[grp == "Grupo A"],
            s=10, alpha=alpha
        )
        ax.scatter(
            X[grp == "Grupo B"], Y[grp == "Grupo B"],
            s=10, alpha=alpha
        )

        ax.text(
            0.5, 0.90,
            "População real",
            transform=ax.transAxes,
            ha="center", va="center",
            fontsize=14, weight="bold"
        )

        ax.text(
            0.5, 0.08,
            "Antes de analisar, parece tudo equilibrado.",
            transform=ax.transAxes,
            ha="center", va="center",
            fontsize=11
        )

    elif frame < 70:
        ax.scatter(
            X[grp == "Grupo A"], Y[grp == "Grupo A"],
            s=8, alpha=0.15
        )
        ax.scatter(
            X[grp == "Grupo B"], Y[grp == "Grupo B"],
            s=8, alpha=0.15
        )

        rect_alpha = min(1, (frame - 35) / 20)
        rect = Rectangle(
            (5.2, 3.5),
            11 - 5.2,
            6.8 - 3.5,
            fill=False,
            linewidth=2.2,
            alpha=rect_alpha
        )
        ax.add_patch(rect)

        ax.text(
            0.5, 0.90,
            "Amostra observada",
            transform=ax.transAxes,
            ha="center", va="center",
            fontsize=14, weight="bold"
        )

        ax.text(
            0.5, 0.08,
            "Nem todo dado disponível representa bem o problema real.",
            transform=ax.transAxes,
            ha="center", va="center",
            fontsize=11
        )

    elif frame < 105:
        ax.scatter(
            X[~sample_mask], Y[~sample_mask],
            s=6, alpha=0.04
        )
        ax.scatter(
            X[sample_mask & (grp == "Grupo A")],
            Y[sample_mask & (grp == "Grupo A")],
            s=12, alpha=0.85
        )
        ax.scatter(
            X[sample_mask & (grp == "Grupo B")],
            Y[sample_mask & (grp == "Grupo B")],
            s=12, alpha=0.85
        )

        rect = Rectangle(
            (5.2, 3.5),
            11 - 5.2,
            6.8 - 3.5,
            fill=False,
            linewidth=2.2
        )
        ax.add_patch(rect)

        ax.text(
            0.5, 0.90,
            "A amostra ficou enviesada",
            transform=ax.transAxes,
            ha="center", va="center",
            fontsize=14, weight="bold"
        )

        ax.text(
            0.5, 0.08,
            "A conclusão pode mudar antes mesmo da modelagem começar.",
            transform=ax.transAxes,
            ha="center", va="center",
            fontsize=11
        )

    else:
        ax.scatter(
            X[~sample_mask], Y[~sample_mask],
            s=5, alpha=0.03
        )
        ax.scatter(
            X[sample_mask & (grp == "Grupo A")],
            Y[sample_mask & (grp == "Grupo A")],
            s=11, alpha=0.8
        )
        ax.scatter(
            X[sample_mask & (grp == "Grupo B")],
            Y[sample_mask & (grp == "Grupo B")],
            s=11, alpha=0.8
        )

        ax.text(
            0.5, 0.80,
            "População: A = {:.0f}% | B = {:.0f}%".format(pop_a * 100, pop_b * 100),
            transform=ax.transAxes,
            ha="center", va="center",
            fontsize=12
        )

        ax.text(
            0.5, 0.73,
            "Amostra:   A = {:.0f}% | B = {:.0f}%".format(sample_a * 100, sample_b * 100),
            transform=ax.transAxes,
            ha="center", va="center",
            fontsize=12, weight="bold"
        )

        ax.text(
            0.5, 0.23,
            "Em ciência de dados, um bom modelo não corrige sozinho",
            transform=ax.transAxes,
            ha="center", va="center",
            fontsize=11
        )

        ax.text(
            0.5, 0.16,
            "um dado que já chegou enviesado.",
            transform=ax.transAxes,
            ha="center", va="center",
            fontsize=12, weight="bold"
        )

        ax.text(
            0.5, 0.07,
            "Viés de amostragem também é problema de negócio.",
            transform=ax.transAxes,
            ha="center", va="center",
            fontsize=11
        )

anim = FuncAnimation(fig, update, frames=140, interval=90, repeat=True)

HTML(anim.to_jshtml())
```
